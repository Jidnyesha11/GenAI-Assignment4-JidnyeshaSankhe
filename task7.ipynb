```python
# Product Dictionary
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

# Get discount percentage
discount = float(input("Enter Discount Percentage: "))

# Open file in write mode
file = open("discount_report.txt", "w")

total_discounted = 0
count = 0

# Write heading
file.write("Discount percentage:"+ str(discount) + "\n")
file.write("Product | Original Price | Discounted Price\n")

# Calculate discounted prices
for product in prices:

    original_price = prices[product]

    discounted_price = original_price - (original_price * discount / 100)

    file.write(product + " | " + str(original_price) + " | " + str(discounted_price) + "\n")

    total_discounted = total_discounted + discounted_price
    count = count + 1

# Optional Summary
average_discounted = total_discounted / count

file.write("\nTotal Items: " + str(count))
file.write("\nAverage Discounted Price: " + str(average_discounted))

file.close()

# Read and display the file
file = open("discount_report.txt", "r")

print(file.read())

file.close()
```
