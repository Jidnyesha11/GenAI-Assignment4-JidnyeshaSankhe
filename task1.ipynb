```python
sales = [1200, 450, 980, 1500, 3000]

with open("sales_data.txt", "w") as file:

    for sale in sales:

        file.write(str(sale) + "\n")

with open("sales_data.txt", "r") as file:

    data = file.read()

    print("Sales Records:")

    print(data)
```

```python
# csv format data

sales = [1200, 450, 980, 1500, 3000]

with open("sales_data.txt", "w") as file:

    for i in range(len(sales)):

        file.write(str(sales[i]))

        if i != len(sales) - 1:

            file.write(",")
            
```
