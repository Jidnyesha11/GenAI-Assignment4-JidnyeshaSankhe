```python
# Read entire file

with open("sales_data.txt", "r") as file:

    data = file.read()

    print("Complete File:")

    print(data)
```

```python
# Read first line

with open("sales_data.txt", "r") as file:

    first_line = file.readline()

    print("First Line:")

    print(first_line)
```

```python
# Read all lines

with open("sales_data.txt", "r") as file:

    lines = file.readlines()

sales = []

for line in lines:

    sales.append(int(line.strip()))

print("Sales List:")

print(sales)
```
